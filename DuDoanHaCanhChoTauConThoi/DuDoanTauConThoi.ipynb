```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/colab/data/dataVisibility.csv')
data.head()
```

```python
inputs = data.drop("Visibility", axis= 1)
inputs.head()
```

```python
data["Visibility"].head()
```

```python
le_error = LabelEncoder()
le_magnitude = LabelEncoder()
le_class= LabelEncoder()
le_wind = LabelEncoder()
lb_visibility = LabelEncoder()
```

```python
inputs['error_n'] = le_error.fit_transform(inputs['Error'])
inputs['magnitude_n'] = le_magnitude.fit_transform(inputs['Magnitude'])
inputs['class_n'] = le_class.fit_transform(inputs['Class'])
inputs['wind_n'] = le_wind.fit_transform(inputs['Wind'])
```

```python
inputs_n = inputs.drop(['Error','Magnitude','Class','Wind'], axis = 'columns')
inputs_n.head()
```

```python
data['Visibility'] = lb_visibility.fit_transform(data['Visibility'])
outputs = data['Visibility']
outputs.head()
```

```python
train_features = inputs_n.iloc[:15]
train_targets = outputs[:15]
test_features = inputs_n.iloc[15:]
test_targets = outputs[15:]
```

```python
model = tree.DecisionTreeClassifier(criterion = 'entropy')
Tree = model.fit(train_features, train_targets)
```

***Test dữ liệu***


```python
prediction = Tree.predict(test_features)
print(prediction)
```

```python
print(np.array(test_targets))
```

***Test độ chính xác của dữ liệu***

```python
print("Độ chính xác của dữ liệu: ", model.score(test_features, test_targets)*100,"%")
```
